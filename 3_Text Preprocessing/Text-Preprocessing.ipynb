```python
import pandas as pd
import numpy as np
pd.set_option('display.max_colwidth', -1)
```

```python
df = pd.read_csv("cleaned_df.csv").drop("Unnamed: 0" , axis = 1)
df
```

```python
df.Category.value_counts()
```

## remove punctuation

```python
from string import punctuation
print(punctuation)
```

```python
text = " fdsfsd fdsf df@ DS FF1 E$$++3 # RFS DF"
```

```python
def remove_punctuation(text):
    punc_free = "".join(i for i in text if i not in punctuation)
    return punc_free

#storing the puntuation free text
df['punc_free']= df['Article Title'].apply(remove_punctuation)
df.head()
```

## LowerCase Text

```python
df['tit_lower']= df['punc_free'].apply(lambda row : row.lower())
df = df[[ 'punc_free' , 'tit_lower' , 'Category']]
df
```

## word tokenizing

```python
from nltk.tokenize import word_tokenize
df['tokenized_sents']= df['tit_lower'].apply(lambda row : word_tokenize(row))
df = df[['tit_lower', 'tokenized_sents' , 'Category']]
df
```

## remove stopwords

```python
from nltk.corpus import stopwords
stopwords = stopwords.words('english')

def remove_stopwords(text):
    stop_free = [i for i in text if i not in stopwords]
    return stop_free

df['no_stopwords'] = df['tokenized_sents'].apply(remove_stopwords)

df = df[['tokenized_sents', 'no_stopwords' , 'Category']]
df
```

## Steeming 

```python
from nltk.stem.porter import PorterStemmer
```

```python
porter_steemer = PorterStemmer()

#defining the function for stemming
def stemming(text):
    stemmed = [porter_steemer.stem(word) for word in text]
    return stemmed

df['stemmed'] = df['no_stopwords'].apply(stemming)

df = df[['no_stopwords', 'stemmed' , 'Category']]
df
```

## Lemmatizing

```python
from nltk.stem import WordNetLemmatizer
```

```python
word_lemmatizer = WordNetLemmatizer()

#defining the function for lemmatization
def lemmatizer(text):
    lemm_text = [word_lemmatizer.lemmatize(word) for word in text]
    return lemm_text

df['lemmatized'] = df['no_stopwords'].apply(lemmatizer)


# comparing the difference between stemming and lemmatization

df= df[['stemmed','lemmatized','Category']]
df
```

# Save the final results

```python
df['Final_text'] = df['lemmatized'].apply(lambda x : ' '.join(x))

df = df[['Final_text' , 'Category']]
df.to_csv('preprocessed_df.csv')
df
```
